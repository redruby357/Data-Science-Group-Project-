```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

THIS NOTEBOOK IS FOR NORTH EAST ONLY

```python
df = pd.read_csv("Dataset_North_East.csv")
df.head(50)
```

HOMELESS VS YEARS

```python
df["Fiscal Year"] = df["Fiscal Year"].astype(str)

plt.figure(figsize=(10,5))
plt.plot(df["Fiscal Year"], df["Homelessness"], marker='o', linewidth=2)
plt.title("Homelessness Over Time")
plt.xlabel("Fiscal Year")
plt.ylabel("Number of Homeless People")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```

ETHICITY VS YEARS

```python
ethnicity_cols = [ "Ethinicity: White", "Ethinicity: Black/ African / Carribean / Black British","Ethinicity: Asian / Asian British","Ethinicity: Mixed / Multiple ethnic groups","Ethinicity: Other ethnic groups","Ethinicity: Not known"]
```

```python
for col in ethnicity_cols:
    df[col] = df[col].astype(str).str.replace(",", "")
    df[col] = pd.to_numeric(df[col])
```

```python
plt.figure(figsize=(12,6))

bottom = None

for col in ethnicity_cols:
    plt.bar(df["Fiscal Year"], df[col], bottom=bottom, label=col)
    if bottom is None:
        bottom = df[col]
    else:
        bottom = bottom + df[col]

plt.title("Ethnicity Composition of Homelessness Over Time")
plt.xlabel("Fiscal Year")
plt.ylabel("Number of People")
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))

for col in ethnicity_cols:
    plt.plot(df["Fiscal Year"], df[col], marker='o', linewidth=2, label=col)

plt.title("Ethnicity Trends in Homelessness Over Time")
plt.xlabel("Fiscal Year")
plt.ylabel("Number of People")
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
```

```python
for col in df.columns:
    if col != "Fiscal Year":  
        df[col] = (
            df[col]
            .astype(str)
            .str.replace(",", "", regex=False)
            .str.replace("\u00A0", "", regex=False)  
            .str.strip()
        )
        df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Homelessness Factors")
plt.show()
```
